```python
from ultralytics import SAM
from ultralytics import FastSAM
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = FastSAM("FastSAM-s.pt")
```

```python
results = model.predict(source="1.jpg")
```

```python
print(results)
```

```python
masks = results[0].masks.data
masks = masks.numpy()
print(masks)
```

```python
image = cv2.imread("1.jpg")
print(image.shape)
```

```python
my_mask = np.zeros((masks.shape[1], masks.shape[2]))

for i in range(masks.shape[0]):
    for j in range(masks.shape[1]):
        for k in range(masks.shape[2]):
            if masks[i][j][k]:
                my_mask[j][k] = i

print(my_mask)
```

```python
plt.imshow(my_mask)
plt.savefig("mask.png")
```

```python
for a in range(1,21):
  results = model.predict(source= str(a)+".jpg")

  masks = results[0].masks.data
  masks = masks.numpy()

  image = cv2.imread(str(a)+".jpg")

  my_mask = np.zeros((masks.shape[1], masks.shape[2]))
  for i in range(masks.shape[0]):
    for j in range(masks.shape[1]):
        for k in range(masks.shape[2]):
            if masks[i][j][k]:
                my_mask[j][k] = i
  plt.imshow(my_mask)

  plt.savefig("mask"+str(a)+".png")

  plt.show()


```
